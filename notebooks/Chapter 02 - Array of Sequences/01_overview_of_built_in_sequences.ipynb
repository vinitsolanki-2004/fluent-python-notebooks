## üß† What Are Sequences?

A **sequence** is a collection of items arranged in a specific order.

Common characteristics:
- Support indexing and slicing
- Preserve insertion order
- Can be iterated over

Python provides multiple built-in sequence types,
each optimized for different scenarios.

## üìö Built-in Sequence Categories

Python sequences can be broadly classified into:

### üîπ Container Sequences
- Store **references** to objects
- Can hold **heterogeneous data**
- Examples: `list`, `tuple`, `deque`

### üîπ Flat Sequences
- Store **values directly** in memory
- More **memory-efficient**
- Examples: `str`, `bytes`, `array.array`

```python
from collections import abc
```

## üîÅ Mutable vs Immutable Sequences

### Immutable Sequences
- Contents cannot be changed
- Examples: `tuple`, `str`, `bytes`
- Safer and hashable

### Mutable Sequences
- Can be modified in place
- Examples: `list`, `bytearray`, `deque`
- More flexible but less safe

```python
issubclass(tuple, abc.Sequence)
```

```python
issubclass(list, abc.MutableSequence)
```
