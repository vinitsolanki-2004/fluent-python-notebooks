## ðŸ“ž Unpacking in Function Calls

The `*` operator can also unpack iterables
directly into function arguments.

This allows dynamic argument passing while keeping code clean.

````python
t = (20, 8)
divmod(*t)
````

````python
quotient, remainder = divmod(*t)
````

````python
quotient
````

````python
remainder
````

## â­ Extended Iterable Unpacking

Using `*` allows capturing multiple elements:

````python
a, b, *rest = range(5)
a, b, rest
````

````python
a, b, *rest = range(3)
a, b, rest
````

````python
a, b, *rest = range(2)
a, b, rest
````

````python
a, *body, c, d = range(5)
a, body, c, d
````

````python
*head, b, c, d = range(5)
head, b, c, d
````

````python
def fun(a, b, c, d, *rest):
    return a, b, c, d, rest
````

````python
fun(*[1, 2], 3, *range(4, 7))
````

````python
*range(4), 4
````

````python
[*range(4), 4]
````

````python
{*range(4), 4, *(5, 6, 7)}
````

## ðŸš« Ignoring Unwanted Values

The underscore `_` is commonly used to ignore values:

````python
def main():
    print(f'{"":15} | {"latitude":>9} | {"longitude":>9}')
    for name, _, _, (lat, lon) in metro_areas:
        if lon <= 0:
            print(f'{name:15} | {lat:9.4f} | {lon:9.4f}')
````

````python
main()
````
