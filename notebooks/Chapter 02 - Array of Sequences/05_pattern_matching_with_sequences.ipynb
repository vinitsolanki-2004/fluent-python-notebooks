## ðŸ§  What Is Pattern Matching?

Pattern matching allows Python to:
- Inspect the **structure** of data
- Match values **by shape**, not just by value
- Destructure sequences declaratively

It is inspired by functional programming languages
but designed in a Pythonic way.

## ðŸ“¦ Why Pattern Match on Sequences?

Sequences are ideal candidates because:
- Order matters
- Length can be matched
- Elements can be unpacked directly

This makes pattern matching expressive and readable.

## ðŸ§ª Basic Sequence Pattern Matching

In the following examples:
- Each `case` describes a **shape**
- Matching stops at the **first successful case**
- Variable names capture matched elements

```python
def handle_command(self, message):
 match message:
    case ['BEEPER', frequency, times]:
        self.beep(times, frequency)
    case ['NECK', angle]:
        self.rotate_neck(angle)
    case ['LED', ident, intensity]:
        self.leds[ident].set_brightness(ident, intensity)
    case ['LED', ident, red, green, blue]:
        self.leds[ident].set_color(ident, red, green, blue)
    case _:
        raise InvalidCommand(message)   
```

## ðŸš« Wildcards and Ignoring Elements

The underscore `_` acts as:
- A wildcard (match anything)
- A way to ignore values safely

This keeps patterns focused on what matters.

```python
def main():
    print(f'{"":15} | {"latitude":>9} | {"longitude":>9}')
    for record in metro_areas:
        match record:
            case [name, _, _, (lat, lon)] if lon <= 0:
                print(f'{name:15} | {lat:9.4f} | {lon:9.4f}')
```
