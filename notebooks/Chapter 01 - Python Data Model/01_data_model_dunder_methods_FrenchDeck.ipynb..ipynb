## ğŸ§  Understanding the Python Data Model

Python objects gain behavior not through inheritance,
but by implementing **special methods** such as:

- `__len__`
- `__getitem__`
- `__repr__`

These methods allow objects to integrate naturally with Python syntax
and standard library functions.

## ğŸ§ª Example: A Pythonic Deck of Cards

We define a `FrenchDeck` class that:

- Supports `len(deck)`
- Allows indexing and slicing
- Works with `random.choice`
- Requires **no custom iterator**

This demonstrates how powerful protocol-based design is in Python.

```python
import collections

Card = collections.namedtuple('Card', ['rank', 'suit'])

class FrenchDeck:
    ranks = [str(n) for n in range(2, 11)] + list('JQKA')
    suits = 'spades diamonds clubs hearts'.split()

    def __init__(self):
        self._cards = [
            Card(rank, suit)
            for suit in self.suits
            for rank in self.ranks
        ]

    def __len__(self):
        return len(self._cards)

    def __getitem__(self, position):
        return self._cards[position]
```

## ğŸ“ Observations

- Implementing `__getitem__` automatically enables iteration
- Python falls back to sequence protocol if no iterator is defined
- Many built-in behaviors come *for free* once protocols are satisfied

```python
for card in deck:
    print(card)
```

```python
for card in reversed(deck):
    print(card)
```

```python
Card('Q', 'hearts') in deck
```

Sorting of Deck

```python
suit_values = dict(spades=3, hearts=2, diamonds=1, clubs=0)
def spades_high(card):
 rank_value = FrenchDeck.ranks.index(card.rank)
 return rank_value * len(suit_values) + suit_values[card.suit]
```

```python
for card in sorted(deck, key=spades_high): 
    print(card)
```
